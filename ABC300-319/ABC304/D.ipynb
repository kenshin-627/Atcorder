```python
from collections import defaultdict

class UnionFind():
    def __init__(self, n):
        self.n = n
        self.parents = [-1] * n

    def find(self, x):
        if self.parents[x] < 0:
            return x
        else:
            self.parents[x] = self.find(self.parents[x])
            return self.parents[x]

    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)

        if x == y:
            return

        self.parents[x] += self.parents[y]
        self.parents[y] = x

    def size(self, x):
        return -self.parents[self.find(x)]

    def same(self, x, y):
        return self.find(x) == self.find(y)

    def members(self, x):
        root = self.find(x)
        return [i for i in range(self.n) if self.find(i) == root]

    def roots(self):
        return [i for i, x in enumerate(self.parents) if x < 0]

    def group_count(self):
        return len(self.roots())

    def all_group_members(self):
        group_members = defaultdict(list)
        for member in range(self.n):
            group_members[self.find(member)].append(member)
        return group_members

def main():
    W,H = map(int,input().split())
    X_list = []
    Y_list = []
    N = int(input())
    for _ in [0]*N:
        p,q = map(int,input().split())
        X_list.append((p,q))
        Y_list.append((q,p))
    X_list.sort()
    Y_list.sort()
    A = int(input())
    a_A = list(map(int,input().split()))
    i = 0
    j = 0
    X_dict = defaultdict(list)
    while j < N:
        if X_list[j][0] < a_A[i]:
            X_dict[i].append(X_list[j])
            j += 1
        else:
            i += 1
            if i == A:
                X_dict[i] = X_list[j:]
                break
    B = int(input())
    b_B = list(map(int,input().split()))
    i = 0
    j = 0
    Y_dict = defaultdict(int)
    while j < N:
        if Y_list[j][0] < b_B[i]:
            Y_dict[Y_list[j]] = i
            j += 1
        else:
            i += 1
            if i == B:
                for t in Y_list[j:]:
                    Y_dict[t] = i
                break
    min_ = 10**9
    max_ = 0
    cnt = 0
    for val in X_dict.values():
        tmp_dict = defaultdict(int)
        for x,y in val:
            tmp_dict[Y_dict[(y,x)]] += 1
        # tmp_dict の最大値と最小値を使って更新する
        vals = tmp_dict.values()
        cnt += len(set(tmp_dict.keys()))
        min_ = min(min_,min(vals))
        max_ = max(max_,max(vals))
    if cnt != (A+1)*(B+1):
        min_ = 0
    print(min_,max_)

if __name__ == "__main__":
    main()
```
