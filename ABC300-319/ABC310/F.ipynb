```python
# from functools import lru_cache
# import sys
# sys.setrecursionlimit(10**9)
# input = sys.stdin.readline
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
import math
import heapq
from collections import OrderedDict
import bisect
from collections import defaultdict

from collections import deque

def main():
    N = int(input())
    A = list(map(int, input().split()))
    mod = 998244353
    # i番目までのサイコロを振った時点で10ができるようになった通り数
    can_ten = [0]*N
    # 全通り数
    all_num = 1
    # q: 今までの出目の和の集合
    q = deque([set([0]), -1])
    # a: 出目の最大値
    for k,a in enumerate(A):
        all_num = all_num * a % mod
        if a > 10:
            a = 10
        while q:
            # s: 今までの出目の和の集合
            s = q.popleft()
            if s == -1:
                q.append(-1)
                break
            # i: 今回の出目
            for i in range(1,a+1):
                # tmp: 今までの出目の和の集合 に 今回の出目を足したもの
                tmp = s.copy()
                # j: 今までの出目の和
                for j in s:
                    if j+i > 10:
                        continue
                    tmp.add(j+i)
                if 10 in tmp:
                    can_ten[k] += 1
                else:
                    q.append(tmp)
    can_ten_num = 0
    tmp_num = 1
    # print(all_num)
    # print(can_ten)
    for i,c in enumerate(reversed(can_ten)):
        can_ten_num = (can_ten_num + c * tmp_num) % mod
        tmp_num = (tmp_num * A[-1-i]) % mod
    # print(can_ten_num)
    print((can_ten_num * pow(tmp_num, mod-2, mod)) % mod)
    
if __name__ == "__main__":
    main()
```
