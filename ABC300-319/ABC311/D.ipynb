```python
# from functools import lru_cache
# import sys
# sys.setrecursionlimit(10**9)
# input = sys.stdin.readline
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
import math
import heapq
from collections import OrderedDict
import bisect
from collections import deque
from collections import defaultdict

# 現在のi,jと，横，縦それぞれの方向の壁に隣接する氷のマスが昇順ソートされているリストを受け取って，次のi,jの候補をリストで返す
def next_cordinate(i, j, h_list_dict, w_list_dict, S):
    adj_list = []
    # 上下方向
    i_index = bisect.bisect_left(h_list_dict[j], i)
    # もし h_list_dict[j][i_index] が i と等しいなら，上下に岩があるか見て，なければ端まで追加
    if i_index < len(h_list_dict[j]):
        if h_list_dict[j][i_index] == i:
            if S[i-1][j] == ".":
                if 0 <= i_index-1:
                    adj_list.append((h_list_dict[j][i_index-1], j))
            if S[i+1][j] == ".":
                if i_index+1 < len(h_list_dict[j]):
                    adj_list.append((h_list_dict[j][i_index+1], j))
        else:
            if 0 <= i_index-1:
                adj_list.append((h_list_dict[j][i_index-1], j))
            if i_index < len(h_list_dict[j]):
                adj_list.append((h_list_dict[j][i_index], j))
    # 左右方向
    j_index = bisect.bisect_left(w_list_dict[i], j)
    if j_index < len(w_list_dict[i]):
        if w_list_dict[i][j_index] == j:
            if S[i][j-1] == ".":
                if 0 <= j_index-1:
                    adj_list.append((i, w_list_dict[i][j_index-1]))
            if S[i][j+1] == ".":
                if j_index+1 < len(w_list_dict[i]):
                    adj_list.append((i, w_list_dict[i][j_index+1]))
        else:
            if 0 <= j_index-1:
                adj_list.append((i, w_list_dict[i][j_index-1]))
            if j_index < len(w_list_dict[i]):
                adj_list.append((i, w_list_dict[i][j_index]))
    return adj_list

def main():
    N,M = map(int, input().split())
    S = []
    for _ in [0]*N:
        S.append(list(input()))
    # 縦方向の壁に隣接する氷のマスを昇順ソートしたリスト
    h_list_dict = defaultdict(list)
    # 横方向の壁に隣接する氷のマスを昇順ソートしたリスト
    w_list_dict = defaultdict(list)
    for i in range(N):
        bef = "#"
        for j in range(M):
            # S[i][j]が氷から壁になったら，S[i][j-1]を登録，S[i][j]が壁から氷になったら，S[i][j]を登録
            if bef == "#" and S[i][j] == ".":
                w_list_dict[i].append(j)
            elif bef == "." and S[i][j] == "#" and S[i][j-2] != "#":
                w_list_dict[i].append(j-1)
            bef = S[i][j]
    for j in range(M):
        bef = "#"
        for i in range(N):
            if bef == "#" and S[i][j] == ".":
                h_list_dict[j].append(i)
            elif bef == "." and S[i][j] == "#" and S[i-2][j] != "#":
                h_list_dict[j].append(i-1)
            bef = S[i][j]
    # 初期位置
    q = deque([(1,1)])
    seen = set([(1,1)])
    while q:
        i, j = q.popleft()
        for i2, j2 in next_cordinate(i, j, h_list_dict, w_list_dict, S):
            if (i2, j2) not in seen:
                q.append((i2, j2))
                # print((i2,j2))
            # i から i2, j から j2 に移動した時の間のます全てをseenに追加
            for i3 in range(min(i, i2), max(i, i2)+1):
                for j3 in range(min(j, j2), max(j, j2)+1):
                    seen.add((i3, j3))
    print(len(seen))
    # print(seen)
    
if __name__ == "__main__":
    main()
```
