```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**9)
# input = sys.stdin.readline
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement
from copy import deepcopy
import math
import heapq
from collections import OrderedDict
import bisect
from collections import defaultdict
from collections import deque

def main():
    H,W = map(int,input().split())
    A = [list(input()) for _ in [0]*H]
    # 壁にする候補
    wall = set()
    # スタートとゴール
    start = tuple()
    goal = tuple()
    # 横方向
    for i in range(H):
        for j in range(W):
            if A[i][j] == ">":
                wall.add((i,j))
                j_ = j+1
                while j_ < W:
                    if A[i][j_] == "." or A[i][j_] == "|":
                        wall.add((i,j_))
                        A[i][j_] = "-"
                    else:
                        break
                    j_ += 1
            elif A[i][j] == "<":
                wall.add((i,j))
                j_ = j-1
                while j_ >= 0:
                    if A[i][j_] == "." or A[i][j_] == "|":
                        wall.add((i,j_))
                        A[i][j_] = "-"
                    else:
                        break
                    j_ -= 1
            elif A[i][j] == "S":
                start = (i,j)
            elif A[i][j] == "G":
                goal = (i,j)
    # 縦方向
    for j in range(W):
        for i in range(H):
            if A[i][j] == "v":
                wall.add((i,j))
                i_ = i+1
                while i_ < H:
                    if A[i_][j] == "." or A[i_][j] == "-":
                        wall.add((i_,j))
                        A[i_][j] = "|"
                    else:
                        break
                    i_ += 1
            elif A[i][j] == "^":
                wall.add((i,j))
                i_ = i-1
                while i_ >= 0:
                    if A[i_][j] == "." or A[i_][j] == "-":
                        wall.add((i_,j))
                        A[i_][j] = "|"
                    else:
                        break
                    i_ -= 1
    for w in wall:
        A[w[0]][w[1]] = "#"
    # スタート地点からBFSでゴールまでの距離を求める
    dist = [[0]*W for _ in [0]*H]
    q = deque([start])
    seen = set([start])
    while q:
        i,j = q.popleft()
        for di,dj in [(1,0),(0,1),(-1,0),(0,-1)]:
            i_,j_ = i+di,j+dj
            if (i_,j_) not in seen:
                if 0 <= i_ < H:
                    if 0 <= j_ < W:
                        if A[i_][j_] != "#":
                            dist[i_][j_] = dist[i][j]+1
                            q.append((i_,j_))
                            seen.add((i_,j_))
    if dist[goal[0]][goal[1]] == 0:
        print(-1)
    else:
        print(dist[goal[0]][goal[1]])
    
if __name__ == "__main__":
    main()
```

```python
from collections import deque

def main():
    H,W = map(int,input().split())
    A = [list(input()) for _ in [0]*H]
    # スタートとゴール
    start = tuple()
    goal = tuple()
    # 各方向を見ている人の位置
    right = set()
    left = set()
    up = set()
    down = set()
    for i in range(H):
        for j in range(W):
            if A[i][j] == ".":
                continue
            if A[i][j] == ">":
                right.add((i,j))
            elif A[i][j] == "^":
                up.add((i,j))
            elif A[i][j] == "<":
                left.add((i,j))
            elif A[i][j] == "v":
                down.add((i,j))
            elif A[i][j] == "S":
                start = (i,j)
            elif A[i][j] == "G":
                goal = (i,j)
    # 右を見ている人の位置から壁にする候補を探す
    for r in right:
        i,j = r
        j += 1
        while j < W:
            if A[i][j] == ".":
                A[i][j] = "-"
            else:
                break
            j += 1
    for l in left:
        i,j = l
        j -= 1
        while j >= 0:
            if A[i][j] == ".":
                A[i][j] = "-"
            else:
                break
            j -= 1
    for d in down:
        i,j = d
        i += 1
        while i < H:
            if A[i][j] == "." or A[i][j] == "-":
                A[i][j] = "|"
            else:
                break
            i += 1
    for u in up:
        i,j = u
        i -= 1
        while i >= 0:
            if A[i][j] == "." or A[i][j] == "-":
                A[i][j] = "|"
            else:
                break
            i -= 1
    # スタート地点からBFSでゴールまでの距離を求める
    dist = [[0]*W for _ in [0]*H]
    q = deque([start])
    while q:
        i,j = q.popleft()
        for di,dj in [(1,0),(0,1),(-1,0),(0,-1)]:
            i_,j_ = i+di,j+dj
            if 0 <= i_ < H:
                if 0 <= j_ < W:
                    if dist[i_][j_] == 0:
                        if A[i_][j_] == ".":
                            dist[i_][j_] = dist[i][j]+1
                            q.append((i_,j_))
                        elif A[i_][j_] == "G":
                            print(dist[i][j]+1)
                            return
    print(-1)
    
if __name__ == "__main__":
    main()
```
