```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**9)
# input = sys.stdin.readline
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement
from itertools import product
from itertools import accumulate
from itertools import groupby
from copy import deepcopy
import math
import heapq
from collections import OrderedDict
import bisect
from collections import deque
from collections import defaultdict

def main():
    N = int(input())
    X = list(map(int, input().split()))
    L = list(map(int, input().split()))
    kaisa_dict = defaultdict(lambda: defaultdict(int))
    for i,x in enumerate(X):
        for l in L:
            kaisa_dict[x-l][i] += 1
            kaisa_dict[x+l+1][i] -= 1
    # 各座標の時に宝石を何個の足で掴むことができるか{-3:[3,1,0]}なら-3の時にX0を3個,X1を1個,X2を0個の足で掴める
    ruiseki_dict = defaultdict(lambda: [0]*N)
    # ruiseki_dictに直前に記録した値
    bef_ruiseki = [0]*N
    # 直前に各要素が[N,N-1,...,1]以上になったときのkeyの値
    ok_key = int()
    # 直前に各要素が[N,N-1,...,1]以上になったかどうか，なったならTrue
    flag = False
    ans = 0
    for key in sorted(kaisa_dict.keys()):
        # key:-16, kaisa_dict[key]: {0:1, 1:0, 2:0}
        for i in kaisa_dict[key].keys():
            bef_ruiseki[i] += kaisa_dict[key][i]
        ruiseki_dict[key] = bef_ruiseki.copy()
        i = N
        tmp = sorted(ruiseki_dict[key])
        while i > 0:
            if tmp[i-1] < i:
                break
            i -= 1
        else:
            if flag:
                continue
            flag = True
            ok_key = key
            continue
        if flag:
            ans += key - ok_key
            flag = False
    print(ans)
    
if __name__ == "__main__":
    main()
```
