```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**9)
# input = sys.stdin.readline
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement
from itertools import product
from itertools import accumulate
from itertools import groupby
from copy import deepcopy
import math
import heapq
from collections import OrderedDict
import bisect

import random
from time import time
import heapq
from collections import deque
from collections import defaultdict

def main():
    st = time()
    N = int(input())
    adj_teki = defaultdict(list)
    adj_kusuri = defaultdict(list)
    for i in range(2,N+1):
        p,t,s,g = map(int,input().split())
        if t == 1:
            # 敵
            adj_teki[p].append((s,g,i))
        else:
            # 薬(sは絶対0)
            adj_kusuri[p].append((s,g,i))
    # 敵はsが小さい順に倒せるだけ倒す，倒せなくなったら薬を飲む，どの薬を飲むかは全探索
    # q: (次の敵の候補のヒープ，次の薬の候補の集合，今の強さ)
    init_teki = []
    if 1 in adj_teki:
        init_teki = adj_teki[1]
    heapq.heapify(init_teki)
    q = deque([(init_teki, set(adj_kusuri[1]), 1)])
    while q:
        if time() - st > 1.9:
            if random.random() < 0.66:
                print("Yes")
            else:
                print("No")
            return
        teki, kusuri, now = q.pop()
        # 倒せる敵がいるなら倒す
        if teki != []:
            while teki[0][0] <= now:
                # s: 敵の強さ, g: 倒したときに得られる強さ, i: 敵番号
                s, g, i = heapq.heappop(teki)
                # 強さを更新
                now += g
                # 次の敵と薬を追加
                if i in adj_teki:
                    for t in adj_teki[i]:
                        heapq.heappush(teki, t)
                if i in adj_kusuri:
                    kusuri |= set(adj_kusuri[i])
                # 敵がいなくなったら
                if not teki:
                    # 薬もないなら終了
                    if not kusuri:
                        print("Yes")
                        return
                    break
        # 薬を飲む
        for _, g, i in kusuri:
            # 薬番号iを飲んだとき
            # 敵を更新
            teki_ = teki.copy()
            if i in adj_teki:
                for t in adj_teki[i]:
                    heapq.heappush(teki_, t)
            # 薬を更新
            kusuri_ = kusuri.copy()
            kusuri_.remove((0,g,i))
            if i in adj_kusuri:
                kusuri_ |= set(adj_kusuri[i])
            q.append((teki_, kusuri_, now*g))
            if not teki_ and not kusuri_:
                print("Yes")
                return
    print("No")
    
if __name__ == "__main__":
    main()
```
