```python
from itertools import combinations
from collections import deque
from collections import defaultdict
from functools import lru_cache

ans = 0

# 10進数を2進数に変換
@lru_cache(maxsize=None)
def dec2bin(x):
    if x == 0:
        return "0"
    ans = ""
    while x > 0:
        ans += str(x%2)
        x //= 2
    return ans[::-1]

def main():
    H,W,T = map(int,input().split())
    A = []
    s = []
    g = []
    snack = []
    for i in range(H):
        Ai = input()
        tmp = []
        for j,aij in enumerate(Ai):
            if aij == "S":
                s.append((i,j))
            elif aij == "G":
                g.append((i,j))
            elif aij == "o":
                snack.append((i,j))
            tmp.append(aij)
        A.append(tmp)
    nodes = s + g + snack
    adj_mat = [[float("inf")] * len(nodes) for _ in range(len(nodes))]
    for n_1,n_2 in combinations(range(len(nodes)), 2):
        i_1,j_1 = nodes[n_1]
        i_2,j_2 = nodes[n_2]
        search_q = deque([(i_1,j_1,0)])
        visited = [[False] * W for _ in range(H)]
        visited[i_1][j_1] = True
        while search_q:
            i,j,d = search_q.popleft()
            if i == i_2 and j == j_2:
                adj_mat[n_1][n_2] = d
                adj_mat[n_2][n_1] = d
                break
            for di,dj in [(1,0),(0,1),(-1,0),(0,-1)]:
                if 0 <= i+di < H and 0 <= j+dj < W:
                    if A[i+di][j+dj] != "#" and visited[i+di][j+dj] == False:
                        search_q.append((i+di,j+dj,d+1))
                        visited[i+di][j+dj] = True
    if adj_mat[0][1] > T:
        print(-1)
        return
    # TSP(goal,S): スタートからS(ビットを整数にして使う)のノードを通ってgoal番目のノードへ向かう最小コスト
    # 巡回セールスマン問題ぽいやつ
    @lru_cache(maxsize=None)
    def TSP(goal,S):
        global ans
        if S == 0:
            if goal != 1:
                if adj_mat[0][goal] + adj_mat[goal][1] <= T:
                    ans = max(ans,1)
            return adj_mat[0][goal]
        tmp = float("inf")
        for i,u in enumerate(reversed(dec2bin(S))):
            if u == "1":
                tmp = min(tmp, TSP(len(nodes)-1-i,S-2**i) + adj_mat[len(nodes)-1-i][goal])
        if goal == 1:
            if tmp <= T:
                ans = max(ans,dec2bin(S).count("1"))
        else:
            if tmp + adj_mat[goal][1] <= T:
                ans = max(ans,dec2bin(S).count("1"))
        return tmp
        
    TSP(1,2**len(snack)-1)
    global ans
    print(ans)

if __name__ == "__main__":
    main()
```
