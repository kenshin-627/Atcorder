```python
def main():
    N,M,D = map(int,input().split())
    A = sorted([*set(map(int,input().split()))])
    B = sorted([*set(map(int,input().split()))])
    N = len(A)
    M = len(B)
    i = 0
    j = 0
    ans = -1
    while True:
        if i+j == N+M-1:
            break
        if i >= N:
            i -= 1
            j += 1
        if j >= M:
            j -= 1
            i += 1
        # A[i] B[j]の差がD以下なら
        if abs(A[i]-B[j]) <= D:
            ans = A[i]+B[j]
        # 次が小さい方のインデックスを進める
        if i == N-1 and j == M-1:
            break
        if i == N-1:
            j += 1
        elif j == M-1:
            i += 1
        else:
            if i != N-1 and j != M-1:
                if A[i+1] < B[j+1]:
                    i += 1
                elif A[i+1] > B[j+1]:
                    j += 1
                else:
                    # 同じ時は前の要素が小さい方を進める
                    if A[i] < B[j]:
                        i += 1
                    else:
                        j += 1
    print(ans)
    
if __name__ == "__main__":
    main()
```
