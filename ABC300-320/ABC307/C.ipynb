```python
from collections import Counter
from itertools import permutations
from itertools import combinations
import math
import heapq
from collections import OrderedDict
import bisect
from collections import deque
from collections import defaultdict

def main():
    HA,WA = map(int,input().split())
    A_black = []
    for i in range(HA):
        A = input()
        for j,a in enumerate(A):
            if a == "#":
                A_black.append((i,j))
    HB,WB = map(int,input().split())
    B_black = []
    for i in range(HB):
        B = input()
        for j,b in enumerate(B):
            if b == "#":
                B_black.append((i,j))
    HX,WX = map(int,input().split())
    X = []
    for i in range(HX):
        X.append(input())
    for i in range(HX+10):
        for j in range(WX+10):
            for k in range(HX+10):
                for l in range(WX+10):
                    # 黒い点の座標を計算する
                    black_set = set()
                    # A_blackを(i,j)だけずらす
                    for a in A_black:
                        black_set.add((a[0]+i,a[1]+j))
                    # B_blackを(k,l)だけずらす
                    for b in B_black:
                        black_set.add((b[0]+k,b[1]+l))
                    # Xの各座標があってるか確認して，一個でも違ったら次行く，全部あってたらOK
                    cnt = 0
                    for x in range(HX):
                        for y in range(WX):
                            if X[x][y] == "#":
                                cnt += 1
                                if (x+10,y+10) not in black_set:
                                    break
                            else:
                                if (x+10,y+10) in black_set:
                                    break
                        else:
                            continue
                        break
                    else:
                        if len(black_set) == cnt:
                            print("Yes")
                            return
    print("No")
    
if __name__ == "__main__":
    main()
```
