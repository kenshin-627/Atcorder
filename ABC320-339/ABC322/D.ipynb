```python
from itertools import product

# N*Mの行列を右(左やった)に90度回転させる
def rotate_right(A):
    N = len(A)
    M = len(A[0])
    A_1 = [[0] * N for _ in range(M)]
    for i in range(N):
        for j in range(M):
            A_1[j][i] = A[i][M-1-j]
    return A_1

def main():
    A = [[0]*4 for _ in range(4)]
    B = [[0]*4 for _ in range(4)]
    C = [[0]*4 for _ in range(4)]
    sum_ = 0
    for i in range(4):
        P = input()
        for j in range(4):
            if P[j] == "#":
                A[i][j] = 1
                sum_ += 1
    for i in range(4):
        P = input()
        for j in range(4):
            if P[j] == "#":
                B[i][j] = 1
                sum_ += 1
    for i in range(4):
        P = input()
        for j in range(4):
            if P[j] == "#":
                C[i][j] = 1
                sum_ += 1
    if sum_ != 16:
        print("No")
        return
    # Aの全ての要素が0の行と列を削除
    A = [A[i] for i in range(4) if sum(A[i]) != 0]
    A = [list(x) for x in zip(*A)]
    A = [A[i] for i in range(4) if sum(A[i]) != 0]
    A = [list(x) for x in zip(*A)]
    # Bの全ての要素が0の行と列を削除
    B = [B[i] for i in range(4) if sum(B[i]) != 0]
    B = [list(x) for x in zip(*B)]
    B = [B[i] for i in range(4) if sum(B[i]) != 0]
    B = [list(x) for x in zip(*B)]
    # Cの全ての要素が0の行と列を削除
    C = [C[i] for i in range(4) if sum(C[i]) != 0]
    C = [list(x) for x in zip(*C)]
    C = [C[i] for i in range(4) if sum(C[i]) != 0]
    C = [list(x) for x in zip(*C)]
    # Aを右に90度回転したもの
    A_1 = rotate_right(A)
    # Aを右に180度回転したもの
    A_2 = rotate_right(A_1)
    # Aを右に270度回転したもの
    A_3 = rotate_right(A_2)
    B_1 = rotate_right(B)
    B_2 = rotate_right(B_1)
    B_3 = rotate_right(B_2)
    C_1 = rotate_right(C)
    C_2 = rotate_right(C_1)
    C_3 = rotate_right(C_2)
    A_ = [A, A_1, A_2, A_3]
    B_ = [B, B_1, B_2, B_3]
    C_ = [C, C_1, C_2, C_3]
    # A_を置く
    for A__ in A_:
        # B_を置く
        for B__ in B_:
            # C_を置く
            for C__ in C_:
                # A_を置く位置
                for i,j in product(range(5-len(A__)), range(5-len(A__[0]))):
                    # B_を置く位置
                    for k,l in product(range(5-len(B__)), range(5-len(B__[0]))):
                        # C_を置く位置
                        for m,n in product(range(5-len(C__)), range(5-len(C__[0]))):
                            tmp = [[0]*4 for _ in range(4)]
                            for x,y in product(range(len(A__)), range(len(A__[0]))):
                                tmp[i+x][j+y] += A__[x][y]
                            # if A__ == A_[3] and B__ == B_[0] and C__ == C_[3]:
                            #     if i == 0 and j == 2 and k == 0 and l == 0 and m == 2 and n == 0:
                            #         for tm in tmp:
                            #             print(tm)
                            for x,y in product(range(len(B__)), range(len(B__[0]))):
                                tmp[k+x][l+y] += B__[x][y]
                            # if A__ == A_[3] and B__ == B_[0] and C__ == C_[3]:
                            #     if i == 0 and j == 2 and k == 0 and l == 0 and m == 2 and n == 0:
                            #         for tm in tmp:
                            #             print(tm)
                            for x,y in product(range(len(C__)), range(len(C__[0]))):
                                tmp[m+x][n+y] += C__[x][y]
                            # if A__ == A_[3] and B__ == B_[0] and C__ == C_[3]:
                            #     if i == 0 and j == 2 and k == 0 and l == 0 and m == 2 and n == 0:
                            #         for tm in tmp:
                            #             print(tm)
                            # 全てのますが1かどうか
                            flag = True
                            for x,y in product(range(4), repeat=2):
                                if tmp[x][y] == 0:
                                    flag = False
                                    break
                            if flag:
                                print("Yes")
                                return
    print("No")
    
if __name__ == "__main__":
    main()
```
