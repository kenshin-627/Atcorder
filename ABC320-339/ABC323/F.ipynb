```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**9)
# input = sys.stdin.readline
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement
from itertools import product
from itertools import accumulate
from itertools import groupby
from itertools import pairwise
from copy import deepcopy
import networkx as nx
import networkx.algorithms as nxa
import numpy as np
import math
from heapq import *
from collections import OrderedDict
from bisect import *
from collections import deque
from collections import defaultdict

def main():
    X1,Y1,X2,Y2,X3,Y3 = map(int,input().split())
    ans = 0
    # Xのプラスは右，Yのプラスは上
    # 一直線に並んでる時だけ処理変える
    if X1 == X2 == X3:
        # 縦に並んでる
        # 1,3が2の上下どっちにあるか
        # 1,2,3，3,2,1，のどっちかなら
        if Y1 < Y2 < Y3 or Y3 < Y2 < Y1:
            ans = abs(Y1-Y3)-1
        else:
            ans = abs(Y1-Y2)+3 + abs(Y2-Y3)
    elif Y1 == Y2 == Y3:
        # 横に並んでる
        # 1,3が2の左右どっちにあるか
        # 1,2,3，3,2,1，のどっちかなら
        if X1 < X2 < X3 or X3 < X2 < X1:
            ans = abs(X1-X3)-1
        else:
            ans = abs(X1-X2)+3 + abs(X2-X3)
    else:
        # 2,3が一直線に並んでるとき
        if X2 == X3:
            if Y2 < Y3:
                # 3が上ならY1から(Y2-1)にいく
                ans = abs(Y1-(Y2-1)) + abs(X1-X2) + abs(Y2-Y3)
            else:
                # 3が下ならY1から(Y2+1)にいく
                ans = abs(Y1-(Y2+1)) + abs(X1-X2) + abs(Y2-Y3)
        elif Y2 == Y3:
            # 3が右ならX1から(X2-1)にいく
            if X2 < X3:
                ans = abs(X1-(X2-1)) + abs(Y1-Y2) + abs(X2-X3)
            else:
                # 3が左ならX1から(X2+1)にいく
                ans = abs(X1-(X2+1)) + abs(Y1-Y2) + abs(X2-X3)
        else:
            tmp_X,tmp_Y = 0,0
            if X2 < X3:
                tmp_X = X2-1
            else:
                tmp_X = X2+1
            if Y2 < Y3:
                tmp_Y = Y2-1
            else:
                tmp_Y = Y2+1
            # X1,Y1からtmp_X,Y2，X2,tmp_Yの近い方にいく
            ans = min(abs(X1-tmp_X)+abs(Y1-Y2), abs(X1-X2)+abs(Y1-tmp_Y)) + 2 + abs(X2-X3)+abs(Y2-Y3)
    print(ans)
    
if __name__ == "__main__":
    main()
```
