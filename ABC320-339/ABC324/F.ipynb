```python
from collections import deque
from collections import defaultdict

def main():
    N,M = map(int,input().split())
    adj_dict = defaultdict(list)
    for _ in [0]*M:
        u,v,b,c = map(int,input().split())
        adj_dict[u].append((v,b,c))
    ans = 0
    # (今の頂点番号，今の美しさの総和，今のコストの総和)
    q = deque([(1,0,0)])
    while q:
        u,b,c = q.popleft()
        for v,b2,c2 in adj_dict[u]:
            if v == N:
                if (b+b2) / (c+c2) > ans:
                    ans = (b+b2) / (c+c2)
            else:
                q.append((v,b+b2,c+c2))
    print(ans)

if __name__ == "__main__":
    main()
```

```python
from time import time
from collections import defaultdict

# 1からNまで Σ(bi - ci*x) >= 0 となる経路が存在するか
def can_go(N, adj_dict, x):
    dp = [-10**8]*(N+1)
    dp[1] = 0
    for i in range(1,N+1):
        for v,b,c in adj_dict[i]:
            dp[v] = max(dp[v], dp[i] + b - c*x)
    return dp[N] >= 0

def main():
    st = time()
    N,M = map(int,input().split())
    adj_dict = defaultdict(list)
    for _ in [0]*M:
        u,v,b,c = map(int,input().split())
        adj_dict[u].append((v,b,c))
    # xは 1/(10**4) ~ 10**4
    l,r = 1/(10**4),10**4
    mid = (l+r)/2
    while time() - st < 4.5:
        if can_go(N, adj_dict, mid):
            l = mid
        else:
            r = mid
        mid = (l+r)/2
    print(mid)

if __name__ == "__main__":
    main()
```
