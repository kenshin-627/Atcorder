```python
from heapq import *

def main():
    # 範囲に入っているもののうち，最も早く範囲外に出るものを選ぶ貪欲
    N = int(input())
    # in_event: [入る時間，出る時間] をもつヒープ
    in_event = []
    # out_event: 今範囲内にいるものそれぞれの出る時間 をもつヒープ
    out_event = []
    for _ in [0]*N:
        T,D = map(int,input().split())
        heappush(in_event,[T,T+D])
    ans = 0
    time = 0
    while in_event:
        # print(f"time: {time}")
        # 今範囲内にいるものがない時は次のin_eventまで時間を進める
        if not out_event:
            t,d = heappop(in_event)
            # print(f"in_pop {t},{d}")
            time = t
            heappush(out_event,d)
            # print(f"out_push {d}")
            while in_event:
                if in_event[0][0] == time:
                    t,d = heappop(in_event)
                    # print(f"in_pop {t},{d}")
                    heappush(out_event,d)
                    # print(f"out_push {d}")
                else:
                    break
        # 範囲内にいるものがある時は，その時間に入るものを全てout_eventに追加する
        else:
            while in_event:
                if in_event[0][0] == time:
                    t,d = heappop(in_event)
                    # print(f"in_pop {t},{d}")
                    heappush(out_event,d)
                    # print(f"out_push {d}")
                else:
                    break
        # 範囲外のものを取り出す
        while out_event:
            if out_event[0] < time:
                heappop(out_event)
                # print("捨てる")
                # print(f"out_pop {heappop(out_event)}")
            else:
                break
        else:
            continue
        # 全部追加したからout_eventから最小のものを取り出して，時間を1進めて，ansを1増やす
        heappop(out_event)
        # print(f"out_pop {heappop(out_event)}")
        time += 1
        ans += 1
    while out_event:
        while out_event:
            if out_event[0] < time:
                heappop(out_event)
                # print(f"out_pop {heappop(out_event)}")
            else:
                break
        if not out_event:
            break
        # 最小のものを取り出して，時間を1進めて，ansを1増やす
        heappop(out_event)
        # print(f"out_pop {heappop(out_event)}")
        time += 1
        ans += 1
    print(ans)
    
if __name__ == "__main__":
    main()
```
