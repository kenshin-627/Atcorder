```python
from heapq import *

# 各都市に行ける最小の時間のうち，すでに電車を使っているものと，電車を使っていないものの最小値を別で管理する．
def main():
    N,A,B,C = map(int,input().split())
    D = [list(map(int,input().split())) for _ in range(N)]
    # 電車を使わずに行ってない都市の集合
    S_car_not = set([i for i in range(N)])
    # 電車を使って行ってない都市の集合
    S_train_not = set([i for i in range(N)])
    # 電車を使わずに行った都市の最小時間
    T_car = [10**18] * N
    # 電車を使って行った都市の最小時間
    T_train = [10**18] * N
    T_car[0] = 0
    # time_city_train: [都市に着く時間, 都市の番号, 電車を使うかどうか(0なら使わない，1なら使う)]
    time_city_train = [[0,0,0]]
    while time_city_train:
        time, city, train = heappop(time_city_train)
        # print(time, city, train)
        if train == 0:
            # 電車を使わずに行く場合
            if T_car[city] != time:
                continue
            S_car_not.remove(city)
            for s in S_car_not:
                # 電車を使わずに行く場合
                if T_car[s] > time + D[city][s]*A:
                    heappush(time_city_train, [time + D[city][s]*A, s, 0])
                    T_car[s] = time + D[city][s]*A
                # 電車を使って行く場合
                if T_train[s] > time + D[city][s]*B+C:
                    heappush(time_city_train, [time + D[city][s]*B+C, s, 1])
                    T_train[s] = time + D[city][s]*B+C
        else:
            # 電車を使って行く場合
            if T_train[city] != time:
                continue
            S_train_not.remove(city)
            for s in S_train_not:
                # 電車を使ってしか行けない
                if T_train[s] > time + D[city][s]*B+C:
                    heappush(time_city_train, [time + D[city][s]*B+C, s, 1])
                    T_train[s] = time + D[city][s]*B+C
    print(min(T_car[-1], T_train[-1]))
    
if __name__ == "__main__":
    main()
```

```python
import networkx as nx

def main():
    N,A,B,C = map(int,input().split())
    S_G = nx.Graph()
    G_S = nx.Graph()
    for i in range(N):
        D = list(map(int,input().split()))
        for j in range(i+1,N):
            S_G.add_edge(i,j,weight=D[j]*A)
            G_S.add_edge(i,j,weight=D[j]*B+C)
    length_S = nx.single_source_dijkstra_path_length(S_G,0)
    length_G = nx.single_source_dijkstra_path_length(G_S,N-1)
    ans = float('inf')
    for i in range(N):
        ans = min(ans, length_S[i]+length_G[i])
    print(ans)
    
if __name__ == "__main__":
    main()
```

```python
from heapq import *
from collections import defaultdict

# 負の辺がないグラフ，始点sから行ける頂点とその距離を辞書で返す
# 隣接辞書のadj_dictは{頂点:[(距離, 隣接頂点), ...]}の形式
def dijkstra(s, adj_dict):
    dist_dict = {s:0}
    q = sorted(adj_dict[s])
    while q:
        dist, v = heappop(q)
        if v in dist_dict:
            continue
        dist_dict[v] = dist
        for d, u in adj_dict[v]:
            if u in dist_dict:
                continue
            heappush(q, (dist+d, u))
    return dist_dict

def main():
    N,A,B,C = map(int,input().split())
    adj_dict_S = defaultdict(list)
    adj_dict_G = defaultdict(list)
    for i in range(N):
        D = list(map(int,input().split()))
        for j in range(i+1,N):
            adj_dict_S[i].append((D[j]*A, j))
            adj_dict_S[j].append((D[j]*A, i))
            adj_dict_G[i].append((D[j]*B+C, j))
            adj_dict_G[j].append((D[j]*B+C, i))
    length_S = dijkstra(0, adj_dict_S)
    length_G = dijkstra(N-1, adj_dict_G)
    ans = float('inf')
    for i in range(N):
        if i in length_S and i in length_G:
            ans = min(ans, length_S[i]+length_G[i])
    print(ans)

if __name__ == "__main__":
    main()
```

```python
from heapq import *
from collections import defaultdict

# 負の辺がないグラフ，始点sから行ける頂点とその距離を辞書で返す
# 隣接辞書のadj_dictは{頂点:[(距離, 隣接頂点), ...]}の形式
def dijkstra(s, adj_dict):
    dist_dict = {s:0}
    q = sorted(adj_dict[s])
    while q:
        dist, v = heappop(q)
        if v in dist_dict:
            continue
        dist_dict[v] = dist
        for d, u in adj_dict[v]:
            if u in dist_dict:
                continue
            heappush(q, (dist+d, u))
    return dist_dict

def main():
    N,A,B,C = map(int,input().split())
    adj_dict = defaultdict(list)
    for i in range(N):
        adj_dict[i].append((0, i+N))
        D = list(map(int,input().split()))
        for j in range(i+1,N):
            adj_dict[i].append((D[j]*A, j))
            adj_dict[j].append((D[j]*A, i))
            adj_dict[i+N].append((D[j]*B+C, j+N))
            adj_dict[j+N].append((D[j]*B+C, i+N))
    length = dijkstra(0, adj_dict)
    print(min(length[N-1], length[2*N-1]))

if __name__ == "__main__":
    main()
```
