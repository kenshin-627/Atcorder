```python
from collections import deque
from itertools import combinations
from collections import defaultdict
import sys
sys.setrecursionlimit(10**9) # 再帰関数の呼び出し制限
import pypyjit
pypyjit.set_param('max_unroll_recursion=-1') # pypyの再帰高速化

N = int(input())
R = input()
C = input()

def dfs(i,jkl,board):
    global N,R,C
    j,k,l = jkl
    # j番目がR[i]
    # k番目がR[i]じゃないabcの0番目
    # l番目がR[i]じゃないabcの1番目
    abc = ["A","B","C"]
    abc.remove(R[i])
    for abc_0,abc_1 in [[0,1],[1,0]]:
        if i == 0:
            # R[0],R[0]じゃないabcの0番目,R[0]じゃないabcの1番目と,
            # C[j],C[k],C[l]が一致するか
            if not (R[0] == C[j] and abc[abc_0] == C[k] and abc[abc_1] == C[l]):
                continue
        # i行目を"."で埋める
        board[i] = ["."]*N
        # 盤面の更新
        board[i][j] = R[i]
        board[i][k] = abc[abc_0]
        board[i][l] = abc[abc_1]
        # N-1行目を更新したら判定する
        if i == N-1:
            # m列目にabcが全て含まれているかつ，m列目の一番上がC[m]であるか
            board_t = list(map(list, zip(*board)))
            for m in range(N):
                col_m = board_t[m]
                # m列目
                dic = defaultdict(int)
                first_abc = ""
                for n in range(N):
                    if col_m[n] != "." and first_abc == "":
                        first_abc = col_m[n]
                    dic[col_m[n]] += 1
                if dic["A"] == 1 and dic["B"] == 1 and dic["C"] == 1 and first_abc == C[m]:
                    continue
                else:
                    break
            else:
                print("Yes")
                for board_row in board:
                    print("".join(board_row))
                exit()
        else:
            # i+1行目の更新
            for jkl_next in list(combinations(range(N), 3)):
                dfs(i+1, jkl_next, board)

# q: [i行目, [j,k,l]]
for jkl in list(combinations(range(N), 3)):
    dfs(0, jkl, [["."]*N for _ in range(N)])
print("No")
```
