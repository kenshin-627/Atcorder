```python
from itertools import combinations
from collections import defaultdict
from more_itertools import distribute

def main():
    N,X,Y = map(int,input().split())
    A = list(map(int,input().split()))
    # Aの偶数番目(0-indexed)だけ集めたものがY，奇数番目だけ集めたものがXになればいい
    A_even_bef, A_odd_bef, A_even_aft, A_odd_aft = map(list, distribute(4,A))
    # 偶数番目の中から選んで足したときに目標とする数
    even_goal = sum(A_even_bef) + sum(A_even_aft) - Y
    if even_goal % 2 == 1:
        print("No")
        exit()
    even_goal = even_goal // 2 + Y
    # 奇数番目の中から選んで足したときに目標とする数
    odd_goal = sum(A_odd_bef) + sum(A_odd_aft) - X
    if odd_goal % 2 == 1:
        print("No")
        exit()
    odd_goal = odd_goal // 2 + X
    # 偶数番目の前半の和で作れる数と，その数を作るときに使った要素
    S_even_bef = defaultdict(tuple)
    S_even_bef[0] = tuple()
    for i in range(1, len(A_even_bef)+1):
        for ps in combinations(A_even_bef, i):
            S_even_bef[sum(ps)] = ps
    # 偶数番目の後半の和で作れる数と，その数を作るときに使った要素
    S_even_aft = defaultdict(tuple)
    S_even_aft[0] = tuple()
    for i in range(1, len(A_even_aft)+1):
        for ps in combinations(A_even_aft, i):
            S_even_aft[sum(ps)] = ps
    # 奇数番目の前半の和で作れる数と，その数を作るときに使った要素
    S_odd_bef = defaultdict(tuple)
    S_odd_bef[0] = tuple()
    for i in range(1, len(A_odd_bef)+1):
        for ps in combinations(A_odd_bef, i):
            S_odd_bef[sum(ps)] = ps
    # 奇数番目の後半の和で作れる数と，その数を作るときに使った要素
    S_odd_aft = defaultdict(tuple)
    S_odd_aft[0] = tuple()
    for i in range(1, len(A_odd_aft)+1):
        for ps in combinations(A_odd_aft, i):
            S_odd_aft[sum(ps)] = ps
    # 偶数番目の前半の和と後半の和で目標とする数を作れるか，作れるなら使った数字と何回使ったか
    ans_even = defaultdict(int)
    for s1 in S_even_bef:
        if s1 == even_goal:
            for a in S_even_bef[s1]:
                ans_even[a] += 1
            break
        if even_goal - s1 in S_even_aft:
            for a in S_even_bef[s1]:
                ans_even[a] += 1
            for a in S_even_aft[even_goal - s1]:
                ans_even[a] += 1
            break
    if ans_even == dict() and even_goal != 0:
        print("No")
        exit()
    # 奇数番目の前半の和と後半の和で目標とする数を作れるか
    ans_odd = defaultdict(int)
    for s1 in S_odd_bef:
        if s1 == odd_goal:
            for a in S_odd_bef[s1]:
                ans_odd[a] += 1
            break
        if odd_goal - s1 in S_odd_aft:
            for a in S_odd_bef[s1]:
                ans_odd[a] += 1
            for a in S_odd_aft[odd_goal - s1]:
                ans_odd[a] += 1
            break
    if ans_odd == dict() and odd_goal != 0:
        print("No")
        exit()
    ans = []
    bef_d = "R"
    for i,a in enumerate(A):
        if i % 2 == 0:
            if ans_even[a] > 0:
                ans_even[a] -= 1
                if bef_d == "R":
                    ans.append("L")
                else:
                    ans.append("R")
                bef_d = "U"
            else:
                if bef_d == "R":
                    ans.append("R")
                else:
                    ans.append("L")
                bef_d = "D"
        else:
            if ans_odd[a] > 0:
                ans_odd[a] -= 1
                if bef_d == "U":
                    ans.append("R")
                else:
                    ans.append("L")
                bef_d = "R"
            else:
                if bef_d == "U":
                    ans.append("L")
                else:
                    ans.append("R")
                bef_d = "L"
    print("Yes")
    print("".join(ans))

if __name__ == "__main__":
    main()
```
