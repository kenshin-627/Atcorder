```python
from collections import defaultdict

# 右端のx座標-左端のx座標 と 上端のy座標-下端のy座標 を比べて大きい方から小さくしていく
# 右端のx座標-左端のx座標 の方が大きい場合は，
# 右端のにある点の数と左端のにある点の数を比べて小さい方からうちに一マス寄せる，という操作を繰り返す

# 実際に点を動かさずに，四角形を小さくしていく
# 今右の範囲外にでてる点の数，左の範囲外にでてる点の数，上の範囲外にでてる点の数，下の範囲外にでてる点の数を管理する
# i行目にある点の数，j列目にある点の数を管理する
def main():
    N,K = map(int,input().split())
    cnt_i = defaultdict(int)
    cnt_j = defaultdict(int)
    r = 0
    l = 10**9
    u = 0
    d = 10**9
    for _ in [0]*N:
        x,y = map(int,input().split())
        cnt_i[y] += 1
        cnt_j[x] += 1
        if r < x:
            r = x
        if x < l:
            l = x
        if u < y:
            u = y
        if y < d:
            d = y
    r_out = cnt_j[r]
    l_out = cnt_j[l]
    u_out = cnt_i[u]
    d_out = cnt_i[d]
    rl_list = sorted(cnt_j.items(), key=lambda x:x[0])
    ud_list = sorted(cnt_i.items(), key=lambda x:x[0])
    r_index = len(rl_list)-1
    l_index = 0
    u_index = len(ud_list)-1
    d_index = 0
    # まず囲った時正方形になるように合わせる
    bef_K = K+1
    while bef_K != K:
        bef_K = K
        if r-l > u-d:
            flag = False
            if r_out < l_out:
                next_r_diff = max(u-d, rl_list[r_index-1][0]-l)
                if next_r_diff == rl_list[r_index-1][0]-l:
                    r_index -= 1
                    flag = True
                next_r = l + next_r_diff
                if (r-next_r)*r_out > K:
                    next_r = r - K//r_out
                    if flag:
                        r_index += 1
                K -= (r-next_r)*r_out
                r = next_r
                r_out += cnt_j[r]
            else:
                next_l_diff = max(u-d, r-rl_list[l_index+1][0])
                if next_l_diff == r-rl_list[l_index+1][0]:
                    l_index += 1
                    flag = True
                next_l = r - next_l_diff
                if (next_l-l)*l_out > K:
                    next_l = l + K//l_out
                    if flag:
                        l_index -= 1
                K -= (next_l-l)*l_out
                l = next_l
                l_out += cnt_j[l]
        else:
            if u_out < d_out:
                next_u_diff = max(r-l, ud_list[u_index-1][0]-d)
                if next_u_diff == ud_list[u_index-1][0]-d:
                    u_index -= 1
                    flag = True
                next_u = d + next_u_diff
                if (u-next_u)*u_out > K:
                    next_u = u - K//u_out
                    if flag:
                        u_index += 1
                K -= (u-next_u)*u_out
                u = next_u
                u_out += cnt_i[u]
            else:
                next_d_diff = max(r-l, u-ud_list[d_index+1][0])
                if next_d_diff == u-ud_list[d_index+1][0]:
                    d_index += 1
                    flag = True
                next_d = u - next_d_diff
                if (next_d-d)*d_out > K:
                    next_d = d + K//d_out
                    if flag:
                        d_index -= 1
                K -= (next_d-d)*d_out
                d = next_d
                d_out += cnt_i[d]
    if r-l != u-d:
        print(max(r-l,u-d))
        return
    # 上下の片方と左右の片方を選んで同時に動かせるだけ動かす
    while K >= 0 and not (r-l == u-d == 0):
        flag_rl = False
        flag_ud = False
        if r_out < l_out and u_out < d_out:
            next_diff = min(rl_list[r_index-1][0]-l, ud_list[u_index-1][0]-d)
            if next_diff == rl_list[r_index-1][0]-l:
                r_index -= 1
                flag_rl = True
            else:
                u_index -= 1
                flag_ud = True
            next_r = l + next_diff
            next_u = d + next_diff
            if (r-next_r)*r_out + (u-next_u)*u_out > K:
                next_r = r - K//(r_out+u_out)
                next_u = u - K//(r_out+u_out)
                if flag_rl:
                    r_index += 1
                else:
                    u_index += 1
            K -= (r-next_r)*r_out + (u-next_u)*u_out
            r = next_r
            u = next_u
            r_out += cnt_j[r]
            u_out += cnt_i[u]
        elif r_out < l_out and u_out >= d_out:
            next_diff = min(rl_list[r_index-1][0]-l, u-ud_list[d_index+1][0])
            if next_diff == rl_list[r_index-1][0]-l:
                r_index -= 1
                flag_rl = True
            else:
                d_index += 1
                flag_ud = True
            next_r = l + next_diff
            next_d = u - next_diff
            if (r-next_r)*r_out + (next_d-d)*d_out > K:
                next_r = r - K//(r_out+d_out)
                next_d = d + K//(r_out+d_out)
                if flag_rl:
                    r_index += 1
                else:
                    d_index -= 1
            K -= (r-next_r)*r_out + (next_d-d)*d_out
            r = next_r
            d = next_d
            r_out += cnt_j[r]
            d_out += cnt_i[d]
        elif r_out >= l_out and u_out < d_out:
            next_diff = min(r-rl_list[l_index+1][0], ud_list[u_index-1][0]-d)
            if next_diff == r-rl_list[l_index+1][0]:
                l_index += 1
                flag_rl = True
            else:
                u_index -= 1
                flag_ud = True
            next_l = r - next_diff
            next_u = d + next_diff
            if (next_l-l)*l_out + (u-next_u)*u_out > K:
                next_l = l + K//(l_out+u_out)
                next_u = u - K//(l_out+u_out)
                if flag_rl:
                    l_index -= 1
                else:
                    u_index += 1
            K -= (next_l-l)*l_out + (u-next_u)*u_out
            l = next_l
            u = next_u
            l_out += cnt_j[l]
            u_out += cnt_i[u]
        else:
            next_diff = min(r-rl_list[l_index+1][0], u-ud_list[d_index+1][0])
            if next_diff == r-rl_list[l_index+1][0]:
                l_index += 1
                flag_rl = True
            else:
                d_index += 1
                flag_ud = True
            next_l = r - next_diff
            next_d = u - next_diff
            if (next_l-l)*l_out + (next_d-d)*d_out > K:
                next_l = l + K//(l_out+d_out)
                next_d = d + K//(l_out+d_out)
                if flag_rl:
                    l_index -= 1
                else:
                    d_index -= 1
            K -= (next_l-l)*l_out + (next_d-d)*d_out
            l = next_l
            d = next_d
            l_out += cnt_j[l]
            d_out += cnt_i[d]
    print(max(r-l,u-d))

if __name__ == "__main__":
    main()
```
