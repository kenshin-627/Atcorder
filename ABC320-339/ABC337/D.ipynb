```python
def main():
    H,W,K = map(int,input().split())
    S = [list(input()) for _ in range(H)]
    # 横方向にK連続で"x"がない領域があるかどうか，あるならその区間の"."の最小値を管理する
    min_dot = float("inf")
    if W >= K:
        for i in range(H):
            # "o","x","."の数
            tmp_list = [0,0,0]
            for j in range(K):
                if S[i][j] == "o":
                    tmp_list[0] += 1
                elif S[i][j] == "x":
                    tmp_list[1] += 1
                else:
                    tmp_list[2] += 1
            if tmp_list[1] == 0 and min_dot > tmp_list[2]:
                min_dot = tmp_list[2]
            for j in range(K,W):
                if S[i][j-K] == "o":
                    tmp_list[0] -= 1
                elif S[i][j-K] == "x":
                    tmp_list[1] -= 1
                else:
                    tmp_list[2] -= 1
                if S[i][j] == "o":
                    tmp_list[0] += 1
                elif S[i][j] == "x":
                    tmp_list[1] += 1
                else:
                    tmp_list[2] += 1
                if tmp_list[1] == 0 and min_dot > tmp_list[2]:
                    min_dot = tmp_list[2]
    # 縦方向にK連続で"x"がない領域があるかどうか，あるならその区間の"."の最小値を管理する
    if H >= K:
        for j in range(W):
            # "o","x","."の数
            tmp_list = [0,0,0]
            for i in range(K):
                if S[i][j] == "o":
                    tmp_list[0] += 1
                elif S[i][j] == "x":
                    tmp_list[1] += 1
                else:
                    tmp_list[2] += 1
            if tmp_list[1] == 0 and min_dot > tmp_list[2]:
                min_dot = tmp_list[2]
            for i in range(K,H):
                if S[i-K][j] == "o":
                    tmp_list[0] -= 1
                elif S[i-K][j] == "x":
                    tmp_list[1] -= 1
                else:
                    tmp_list[2] -= 1
                if S[i][j] == "o":
                    tmp_list[0] += 1
                elif S[i][j] == "x":
                    tmp_list[1] += 1
                else:
                    tmp_list[2] += 1
                if tmp_list[1] == 0 and min_dot > tmp_list[2]:
                    min_dot = tmp_list[2]
    if min_dot == float("inf"):
        print(-1)
    else:
        print(min_dot)

if __name__ == "__main__":
    main()
```
