```python
def main():
    N = int(input())
    Q = list(map(int, input().split()))
    A = list(map(int, input().split()))
    B = list(map(int, input().split()))
    cnt_A = 0
    cnt_B = 0
    while True:
        # QからAを引けるなら引く
        for i in range(N):
            if Q[i] < A[i]:
                break
        else:
            for i in range(N):
                Q[i] -= A[i]
            cnt_A += 1
            continue
        break
    while True:
        # QからBを引けるなら引く
        for i in range(N):
            if Q[i] < B[i]:
                break
        else:
            for i in range(N):
                Q[i] -= B[i]
            cnt_B += 1
            continue
        break
    # Aをcnt_A個作れた
    cnt_max = cnt_A + cnt_B
    # Aを壊してBを作る
    while cnt_A > 0:
        cnt_A -= 1
        for i in range(N):
            Q[i] += A[i]
        while True:
            for i in range(N):
                if Q[i] < B[i]:
                    break
            else:
                for i in range(N):
                    Q[i] -= B[i]
                cnt_B += 1
                continue
            break
        if cnt_max < cnt_A + cnt_B:
            cnt_max = cnt_A + cnt_B
    print(cnt_max)

if __name__ == "__main__":
    main()
```
