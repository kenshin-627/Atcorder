```python
from collections import deque

def main():
    N = int(input())
    S = []
    # i,jにプレイヤー1，k,lにプレイヤー2がいる状態への最小手数
    now_dist = [[[[-1]*N for _ in range(N)] for _ in range(N)] for _ in range(N)]
    i,j,k,l = -1,-1,-1,-1
    for m in range(N):
        s = []
        for n,s_ in enumerate(input()):
            if s_ == "P":
                if i == -1 and j == -1:
                    i,j = m,n
                else:
                    k,l = m,n
                s.append(".")
            else:
                s.append(s_)
        S.append(s)
    now_dist[i][j][k][l] = 0
    q = deque([(i,j,k,l)])
    while q:
        i,j,k,l = q.popleft()
        dist = now_dist[i][j][k][l] + 1
        # 上
        i_,k_ = i,k
        if i > 0 and S[i-1][j] == ".":
            i_ -= 1
        if k > 0 and S[k-1][l] == ".":
            k_ -= 1
        if now_dist[i_][j][k_][l] == -1:
            now_dist[i_][j][k_][l] = dist
            if i_ == k_ and j == l:
                print(dist)
                return
            q.append((i_,j,k_,l))
        # 下
        i_,k_ = i,k
        if i < N-1 and S[i+1][j] == ".":
            i_ += 1
        if k < N-1 and S[k+1][l] == ".":
            k_ += 1
        if now_dist[i_][j][k_][l] == -1:
            now_dist[i_][j][k_][l] = dist
            if i_ == k_ and j == l:
                print(dist)
                return
            q.append((i_,j,k_,l))
        # 左
        j_,l_ = j,l
        if j > 0 and S[i][j-1] == ".":
            j_ -= 1
        if l > 0 and S[k][l-1] == ".":
            l_ -= 1
        if now_dist[i][j_][k][l_] == -1:
            now_dist[i][j_][k][l_] = dist
            if i == k and j_ == l_:
                print(dist)
                return
            q.append((i,j_,k,l_))
        # 右
        j_,l_ = j,l
        if j < N-1 and S[i][j+1] == ".":
            j_ += 1
        if l < N-1 and S[k][l+1] == ".":
            l_ += 1
        if now_dist[i][j_][k][l_] == -1:
            now_dist[i][j_][k][l_] = dist
            if i == k and j_ == l_:
                print(dist)
                return
            q.append((i,j_,k,l_))
    print(-1)

if __name__ == "__main__":
    main()
```
