```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**9)
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
import math
import heapq
from collections import OrderedDict
import bisect
from collections import deque
from collections import defaultdict

import bisect

def main():
    N,M = map(int,input().split())
    A = list(map(int,input().split()))
    lower = A[0]
    upper = A[1]
    A_sorted = sorted(A[2:])
    lower_ind = bisect.bisect_left(A_sorted,lower)
    upper_ind = bisect.bisect_right(A_sorted,upper)
    # lowerより小さい数字の配列
    lower_list = A_sorted[:lower_ind]
    # upperより大きい数字の配列
    upper_list = A_sorted[upper_ind:]
    # print(lower_list)
    # print(upper_list)
    # print(upper_ind-lower_ind)
    deff_lower = []
    deff_upper = []
    for i,l in enumerate(lower_list):
        if i == len(lower_list)-1:
            deff_lower.append(lower-l)
        else:
            deff_lower.append(lower_list[i+1]-l)
    for i,u in enumerate(upper_list):
        if i == 0:
            deff_upper.append(u-upper)
        else:
            deff_upper.append(u-upper_list[i-1])
    # print(deff_lower)
    # print(deff_upper)
    sum_lower = []
    sum_upper = []
    if deff_lower != []:
        sum_lower = [deff_lower[-1]]
        for i in range(2,len(deff_lower)+1):
            sum_lower.append(sum_lower[-1]+deff_lower[-i])
    if deff_upper != []:
        sum_upper = [deff_upper[0]]
        for i in range(1,len(deff_upper)):
            sum_upper.append(sum_upper[-1]+deff_upper[i])
    # print(sum_lower)
    # print(sum_upper)
    if upper_ind-lower_ind >= M:
        print(0)
        return
    ans = float("inf")
    if upper_ind-lower_ind >= 0:
        # 残りの選ぶ数
        rest = M - (upper_ind-lower_ind)
        # iはsum_lowerから選ぶ数
        for i in range(rest+1):
            if i == 0:
                if len(sum_upper) >= rest:
                    ans = min(ans,sum_upper[rest-1])
            elif i == rest:
                if len(sum_lower) >= rest:
                    ans = min(ans,sum_lower[rest-1])
            else:
                if len(sum_lower) >= i and len(sum_upper) >= rest-i:
                    ans = min(ans,sum_lower[i-1]+sum_upper[rest-i-1])
    else:
        # lower_listとupper_listで被ってる個数
        coverd_num = lower_ind - upper_ind
        # 残りの選ぶ数
        rest = M
        # iはsum_lowerから選ぶ数
        for i in range(coverd_num + rest + 1):
            if i == 0:
                if len(sum_upper) >= coverd_num + rest:
                    ans = min(ans,sum_upper[coverd_num + rest - 1])
            elif i == coverd_num + rest:
                if len(sum_lower) >= coverd_num + rest:
                    ans = min(ans,sum_lower[coverd_num + rest - 1])
            else:
                if len(sum_lower) >= i and len(sum_upper) >= coverd_num + rest - i:
                    ans = min(ans,sum_lower[i-1]+sum_upper[coverd_num + rest - i-1])
    print(ans)
    
if __name__ == "__main__":
    main()
```
