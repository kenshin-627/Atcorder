```python
from collections import deque

def main():
    N = int(input())
    S = input()
    T = input()
    if S == T:
        print(0)
        return
    # 最も左にあるBにしないとあかんindex
    left = float("inf")
    # 最も右にあるAにしないとあかんindex
    right = -float("inf")
    # Bにしないとあかんindex
    A2B_list = deque()
    # Aにしないとあかん数
    B2A_list = deque()
    # Tの最も左にあるAのindex
    left_A = -1
    # Tの最も右にあるBのindex
    right_B = -1
    for i in range(N):
        if T[i] == "A" and left_A == -1:
            left_A = i
        if T[i] == "B":
            right_B = i
        if T[i] == "B" and S[i] == "A":
            A2B_list.append(i)
            if left == float("inf"):
                left = i
        if T[i] == "A" and S[i] == "B":
            B2A_list.append(i)
            right = i
    # print(A2B_list)
    # print(B2A_list)
    if left_A == -1 or right_B == -1:
        print(-1)
        return
    # print(left_A, right_B)
    # print(left, right)
    if left_A < left and right < right_B:
        cnt = 0
        while True:
            if len(A2B_list) == 0 or len(B2A_list) == 0:
                break
            A2B = A2B_list.popleft()
            B2A = B2A_list.popleft()
            if B2A < A2B:
                cnt += 1
                continue
            B2A_list.appendleft(B2A)
            cnt += 1
        cnt += len(A2B_list)
        cnt += len(B2A_list)
        print(cnt)
    else:
        print(-1)

if __name__ == "__main__":
    main()
```
