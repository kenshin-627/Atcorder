```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**9)
# input = sys.stdin.readline
from decimal import Decimal
from functools import cmp_to_key
from collections import Counter
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement
from copy import deepcopy
import math
import heapq
from collections import OrderedDict
import bisect
from collections import deque

from itertools import combinations
from collections import defaultdict

def main():
    N = int(input())
    A = list(map(lambda x: int(x)-1, input().split()))
    B = sorted(A)
    # iから出る辺, key: iから出る辺の先j, value: iからjへ出る辺の数
    edge_dict = [defaultdict(int) for _ in range(4)]
    for i in range(N):
        edge_dict[A[i]][B[i]] += 1
    ans = 0
    # 長さ2のサイクルの数を数える
    for i,j in combinations(range(4), 2):
        tmp = min(edge_dict[i][j], edge_dict[j][i])
        ans += tmp
        edge_dict[i][j] -= tmp
        edge_dict[j][i] -= tmp
    # 長さ3のサイクルの数を数える
    for i,j,k in combinations(range(4), 3):
        tmp = min(edge_dict[i][j], edge_dict[j][k], edge_dict[k][i])
        ans += tmp*2
        edge_dict[i][j] -= tmp
        edge_dict[j][k] -= tmp
        edge_dict[k][i] -= tmp
        # 逆回り
        tmp = min(edge_dict[i][k], edge_dict[k][j], edge_dict[j][i])
        ans += tmp*2
        edge_dict[i][k] -= tmp
        edge_dict[k][j] -= tmp
        edge_dict[j][i] -= tmp
    for i,j,k,l in combinations(range(4), 4):
        # 回る順番
        order = [[i,j,k,l],[i,j,l,k],[i,k,j,l],[i,k,l,j],[i,l,j,k],[i,l,k,j]]
        for o in order:
            tmp = min(edge_dict[o[0]][o[1]], edge_dict[o[1]][o[2]], edge_dict[o[2]][o[3]], edge_dict[o[3]][o[0]])
            ans += tmp*3
            edge_dict[o[0]][o[1]] -= tmp
            edge_dict[o[1]][o[2]] -= tmp
            edge_dict[o[2]][o[3]] -= tmp
            edge_dict[o[3]][o[0]] -= tmp
    print(ans)
    
if __name__ == "__main__":
    main()
```
